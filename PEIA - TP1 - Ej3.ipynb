```python
# Se importan las librerías necesarias para este ejercicio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importar el archivo de Excel
df_ventas = pd.read_excel('Datos_primer_TP_20Co2025_a2006.xlsx')
df_ventas
```

```python
# Se observa el tipo de datos de cada columna y que no hay valores nulos
df_ventas.info()
```

```python
# Se observan algunos valores importantes como cantidad de datos, promedio, desviación estándar, 
# mínimo y máximo de cada columna
df_ventas.describe()
```

```python
# Gráfico de todos los datos completos de Ventas en función de Fecha
x_values = df_ventas['Fecha']
y_values = df_ventas['Ventas']

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values)

# Etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.title('Gráfico de Ventas en función de Fecha')
plt.grid()

# Mostrar gráfico
plt.show()
```

```python
# Separo los datos en 3 Dataframe según el año
df_ventas_2021=df_ventas[df_ventas['Fecha'].dt.year==2021]
df_ventas_2021
```

```python
df_ventas_2022=df_ventas[df_ventas['Fecha'].dt.year==2022]
df_ventas_2022
```

```python
df_ventas_2023=df_ventas[df_ventas['Fecha'].dt.year==2023]
df_ventas_2023
```

```python
# Obtengo la función empírica de distribución (ECDF) para cada año
sns.ecdfplot(df_ventas_2021['Ventas'], label='2021', color='blue')  
sns.ecdfplot(df_ventas_2022['Ventas'], label='2022', color='red')
sns.ecdfplot(df_ventas_2023['Ventas'], label='2023', color='green')
plt.legend()
plt.grid()
```

```python
# Obtengo una aproximación a la función de densidad de las ventas (KDE) para cada año
sns.kdeplot(df_ventas_2021['Ventas'], color='blue', linestyle='--', label='2021')
sns.kdeplot(df_ventas_2022['Ventas'], color='red', linestyle='--', label='2022')      
sns.kdeplot(df_ventas_2023['Ventas'], color='green', linestyle='--', label='2023')
plt.legend()
plt.grid()
```
