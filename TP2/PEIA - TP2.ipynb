```python
# Se importan las librerías necesarias para este ejercicio
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([3, 7, 14, 18, 20])
X
```

```python
Y=np.array([4, 54, 194, 283, 465])
Y
```

```python
sumX=np.sum(X)
print(sumX)
```

```python
A=np.array([[10, 124, 1956], [124, 1956, 33892], [1956, 33892, 611748]])
A
```

```python
b=np.array([2000, 35000, 636796])
b
```

```python
xsol=np.linalg.solve(A, b)
xsol
```

```python
# Se grafican los valores experimentales y los valores obtenidos a partir de la solución del sistema de ecuaciones
x_values = np.linspace(0, 25, 1000)  
y_values = xsol[0] + xsol[1] * x_values + xsol[2] * x_values**2  # y = a + b*x + c*x^2

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(X, Y, label=r'Datos experimentales', color='blue', marker='*', linestyle='None')
plt.plot(x_values, y_values, label=r'Modelo obtenido', color='red')

# Etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de datos experimentales vs. modelo obtenido')
plt.legend()
plt.grid()

# Mostrar gráfico
plt.show()
```
