```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de x entre 0 y 1
x_values = np.linspace(-0.2, 0.2, 400)
y1_values = 100 * x_values**2  # y = 100x^2
y2_values = 10 * x_values      # y = 10x

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(x_values, y1_values, label=r'$y = 100x^2$', color='blue')
plt.plot(x_values, y2_values, label=r'$y = 10x$', color='red')

# Etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $y = 100x^2$ y $y = 10x$')
plt.legend()
plt.grid()

# Mostrar gráfico
plt.show()
```
